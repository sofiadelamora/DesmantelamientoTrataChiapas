```python
import networkx as nx
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
```

```python
N =nx.read_edgelist("Datos.txt", create_using = nx.Graph(), nodetype=int)
import xlrd
di = {}
wk = xlrd.open_workbook('labels.xls')
hoja = wk.sheet_by_index(0)   
for i in range(34):
    nodo = hoja.cell(i,0).value
    clave = hoja.cell(i,1).value
    di[nodo] = clave
N = nx.relabel_nodes(N, di)
sorted(N)
print(nx.info(N))
nx.draw(N, with_labels=1)
plt.show()
```

```python
#Mtricas
```

```python
##Grado de cada nodo
gradoC=nx.degree(N)
##Excentricidad
ecc = nx.eccentricity(N)
##Centralidad de cercana
cercania = nx.closeness_centrality(N)
##Centralidad de intermediacin
intermediacion = nx.betweenness_centrality(N)
##Coeficiente de Clustering
clustering = nx.clustering(N) ###the clustering of a node u is the fraction of possible triangles through that node that exist
##Centralidad de eigenvector
eigenvector = nx.eigenvector_centrality(N)
##Densidad
densidad = nx.density(N)
```

```python
##Distribucin de los grados
def dist_grados(N):
    mis_grados = nx.degree(N)
    grados = [v for k, v in mis_grados]
    unicos = list(set(grados))
    
    numero_grados=[]
    
    for i in unicos:
        conteo = grados.count(i)
        numero_grados.append(conteo)
    plt.plot(unicos, numero_grados)
    plt.show
dist_grados(N)
```

```python
##Excentricidad
distribuir_nodos = nx.spring_layout(N)
color = [1000.0 * N.degree(v) for v in N]
tamao = [v * 500 for v in ecc.values()]
plt.figure(figsize=(15,15))
nx.draw_networkx(N, pos=distribuir_nodos , with_labels=1,node_color=color, node_size=tamao)
plt.axis('off')
sorted(ecc, key=ecc.get, reverse=1)
```
